```python
countries = drinking_water["Location"].tolist()
countries = set(countries)
countries = list(countries)
countries
```

```python
# Save config information.
url = "https://restcountries.eu/rest/v2/name/"

# set up lists to hold reponse info
name = []
code = []
rejects = []
region = []
i = 0
```

```python
# Loop through the list of countries and perform a request for data on each 
for country in countries:
    response = requests.get(url + country).json()
    try:
        name.append(response[0]['name'])
        region.append(response[0]['region'])
        code.append(response[0]['numericCode'])
        i = i + 1
    except NameError:
        print(f"Oops - I can't find data on {country}, I'll have to skip it")
        rejects.append(country)
        pass
    except KeyError:
        print(f"Oops - I can't find data on {country}, I'll have to skip it")
        rejects.append(country)
        pass
```
